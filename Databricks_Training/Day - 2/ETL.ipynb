```python
df=spark.read.csv("/Volumes/acc_dev/naval/raw/financial_dataset.csv",header=True,inferSchema=True)
```

```python
df=df.withColumnsRenamed({"CustomerID":"customer_id","TransactionID":"transaction_id","AccountType":"account_type","TransactionType":"transaction_type","AmountUSD":"amount_usd","Branch":"branch","TransactionDate":"transaction_date"})
```

```python
from pyspark.sql.functions import col,current_timestamp,upper
```

```python
df=df.withColumn("ingestion_date", current_timestamp()).withColumn("input_path", lit("/Volumes/acc_dev/naval/raw/financial_dataset.csv"))
```

```python

df.write.mode("overwrite").saveAsTable("acc_dev.bhanu.df_finances")
```
